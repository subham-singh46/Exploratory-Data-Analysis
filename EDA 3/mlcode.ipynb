```python
import pandas as pd
data=pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width','petal_length','petal_width','species'])

data

```

```python
#BASIC SCATTER PLOT

import pandas as pd

import matplotlib.pyplot as plt

plt.plot(data.sepal_length,data.sepal_width, LS='',Marker='o',label='sepal')
```

```python
# MULTIPLE LAYER SCATTER PLOT

plt.plot(data.sepal_length,data.sepal_width,LS='',Marker='o',label='SEPAL')


plt.plot(data.petal_length,data.petal_width,LS='',Marker='o',label='PETAL')
```

```python
#PLOTTING HISTOGRAM

plt.hist(data.sepal_length,bins=25,color='red')
```

```python
#CUSTOMIZING PLOTS

#PLOTTING BAR PLOT AND EDITING X-LABELS AN Y-LABELS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()

ax.barh(np.arange(11),data.sepal_width.iloc[:11])

#SET POSITION OF TICKS AND TICKS LABEL

ax.set_yticks(np.arange(0.4,11.4,1.0))

ax.set_ylabel(np.arange(1,11))

ax.set(xlabel='xlabel',ylabel='ylabel',title='Title')

```

```python
#PANDAS DATAFRAME METHOD
data.groupby('species').mean().plot(color=['red','blue','black','green'],fontsize=10.0,figsize=(4,4))
```

```python
import seaborn as sns
iris=sns.load_dataset('iris')

sns.pairplot(iris,hue='species',size=3,palette='rainbow')
```

```python
#SEABORN HEXBIN PLOT

sns.jointplot(x=iris['sepal_length'],y=iris['sepal_width'],kind='hex')
```

```python
#SEABORN PLOT,FACET GRID

#FIRST PLOT STATEMENT
plot=sns.FacetGrid(iris,col='species',margin_titles=True)
plot.map(plt.hist,'sepal_width',color='green')


#SECOND PLOT STATEMENT
plot=sns.FacetGrid(iris,col='species',margin_titles=True)
plot.map(plt.hist,'sepal_length',color='blue')


```
